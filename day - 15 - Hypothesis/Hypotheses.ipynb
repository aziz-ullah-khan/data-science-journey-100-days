# 2. Hypotheses
---

A hypothesis in simple words is "an idea that can be tested" and it's a supposition with limited knowledge that need to be investigated further. 


**Null hypothesis (H0)** \
A hypothesis to be tested. It's like an innocent until proven guilty. Every Data Scientist try to reject the null hypothesis if there is enough evidence in rejection favor. 

**Alternative hypothesis (H1 or HA)** \
Alternate hypothesis is the counter hypothesis which is contesting with the null hypothesis. Data Scientists always try to provide alternative hypothesis and reject the null hypothesis. 


**Example** \
You are provided with the info that Data Scientists earn \$150k per year, it's a **null hypothesis** for you and as a data scientist you will try to reject the null hypothesis with your **alternate hypothesis** as "data scientist are earning greater than $150k per year. So, you will perform some tests and will rejet the null hypothesis.

# 3. Decisions
---

Decision may be to **reject** the null hypothesis or **accept** the null hypothesis. 

**When to reject the null hypothesis?** \
There is enough statistical evidences on basis of which we are rejecting the null hypothesis that the null hypothesis is not the representation of the truth. 

**When to accept the null hypothesis?** \
When there is no enough data/evidence to support the alternate hypothesis, then null hypothesis will be accepted.

# 5. P-value
---
The p-value is the smallest level of significance at which we can still reject the null hypothesis.

**Ideal P-value** is 0.000 and the cut-off is 0.05. If the p-value is greater than 0.05 then will accept the null hypothesis otherwise reject the null hypothesis.
