**Estimate-** \
Estimate is the ouput of an estimator, and are of two types:
1. Point Estimates \
    A single value estimation like 1, 2.4 ..
2. Confidence Interval estimates\
    Estimating an interval like (1, 30), (100, 200) ..


Note: Confidence intervals are more precise and hence using mostly.

# 2. Confidence Intervals and the Margin of Error
---

**Confidence Interval** is an interval within which we are confident that the population parameters will fall. 


 
$(1-Œ±)$ is the level of confidence. We are $(1-Œ±)*100%$ confident that the population parameter will fall in the specified interval.

Common alphas are: 0.01, 0.05, 0.1.

$[\overline{x} - \text{ME}, \, \overline{x} + \text{ME}]$

$ME = \text{reliability factor} \times \frac{\text{standard deviation}}{\sqrt{\text{sample size}}}$
