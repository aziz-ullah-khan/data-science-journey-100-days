# 1. Types of data
---

There are two types of data; categorical and numerical.

**1. Categorical Data:**
Categrocial data represent groups or categories. 

**2. Numerical data:**
Numerical data represent numbers and may be discrete or continuous. 

**2.1 Discrete:**
Data which is finite to count like number of balls. 

**2.2 Continuous:**
Continuous is inifinite and imppossible to count like temperature.

# 2. Levels of measurement
---

There are two levels of measurement i.e qualitative and quantitative.

**1. Qualitative:**
Qualitative may be Nominal or Ordinal. 

**1.1 Nominal:**
Categories without inherent order.\
Examples: Gender, colors.

**1.2 Ordinal Level:**
Meaningful order, uneven intervals.\
Examples: Education levels.

**2 Quantitative:** \
**2.1 Interval Level:**
Meaningful order, uniform intervals, no true zero.\
Examples: Temperature (Celsius, Fahrenheit).

**2.2 Ratio Level:**
Meaningful order, uniform intervals, true zero.\
Examples: Height, weight, income.
